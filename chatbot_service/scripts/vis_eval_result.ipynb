```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

simple = {
  "Faithfulness": 0.3456,
  "Answer Relevance": 0.4223,
  "Precision@K": 0.1285,
  "Recall@K": 0.3883
}

hyde = {
  "Faithfulness": 0.3085,
  "Answer Relevance": 0.293,
  "Precision@K": 0.0033,
  "Recall@K": 0.01
}

df = pd.DataFrame([
    {"method": "Simple RAG", **simple},
    {"method": "HyDE RAG", **hyde},
])

df_melt = df.melt(id_vars="method", var_name="metric", value_name="score")

plt.figure(figsize=(10,6))
sns.barplot(data=df_melt, x="metric", y="score", hue="method", palette="viridis")
plt.title("So sánh hiệu quả Simple RAG vs HyDE RAG")
plt.ylabel("Score")
plt.xlabel("Metric")
plt.ylim(0, 1)
plt.legend(title="Phương pháp")
plt.show()
```

```python
import numpy as np
import sys

# 1. Tải file .npz
du_lieu = np.load('tfidf_matrix.npz')

# 2. Xem danh sách các mảng (array) đang có trong file này
print("Các keys có trong file:", du_lieu.files)

# 3. Cấu hình để in ĐẦY ĐỦ nội dung ma trận (không bị rút gọn bằng dấu ...)
# 'threshold=sys.maxsize' nghĩa là cho phép in kích thước tối đa
np.set_printoptions(threshold=sys.maxsize)

# 4. Lặp qua từng mảng và in ra
for key in du_lieu.files:
    matrix = du_lieu[key]
    print(f"\n--- Nội dung của mảng '{key}' ---")
    print(matrix)

# 5. Đóng file sau khi sử dụng (tốt cho quản lý bộ nhớ)
du_lieu.close()
```

```python
import numpy as np
import scipy.sparse as sp
import sys

# 1. Load file
loader = np.load('tfidf_matrix.npz')

# 2. Lấy các thành phần cấu tạo nên ma trận thưa
data = loader['data']
indices = loader['indices']
indptr = loader['indptr']
shape = loader['shape']

# 3. Tái tạo lại ma trận thưa (CSR Matrix)
# Lưu ý: TF-IDF thường là CSR, nhưng nếu lỗi bạn có thể thử đổi thành sp.csc_matrix
sparse_matrix = sp.csr_matrix((data, indices, indptr), shape=shape)

# 4. Chuyển sang dạng ma trận đầy đủ (Dense) để có thể in ra nhìn thấy được
# CẢNH BÁO: Nếu ma trận quá lớn (vài chục nghìn dòng/cột), bước này có thể gây tràn RAM!
dense_matrix = sparse_matrix.toarray()

# 5. Cấu hình in đầy đủ
np.set_printoptions(threshold=sys.maxsize, linewidth=200)

print("Hình dạng ma trận:", dense_matrix.shape)
print("Nội dung ma trận đầy đủ:")
print(dense_matrix)
```

```python
import pandas as pd

# (Tiếp tục từ biến sparse_matrix ở trên)
# Chuyển sang DataFrame
df = pd.DataFrame.sparse.from_spmatrix(sparse_matrix)

# Cấu hình in
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

print(df)
```
